```python
#imports

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from bioproc import functions as fn
from bioproc import emg as em
```

```python
#generate sinewave with noise
sine = fn.sinenoise(fs=1000)

#generate EMG signal
emg = em.emgsig()
```

```python
#filtering
emg_filter = fn.iir(emg, plot='N')

#zero padding
emg_pad = em.padding(emg_filter)

#FFT of sinewave
fft = fn.fft(emg_pad)
```

```python
#auto-correlation
acorr = fn.acorr(sine)
```

```python
#rectification of EMG 

emg_rect = fn.rectify(emg_filter)
```

```python
#EMG onsets
emg = np.loadtxt(r".\bioproc\Examples\emg.txt")
emg_onset = em.find_onset(emg)
```

```python
#RMS

emg_rms = fn.rms_sig(emg_rect, 20)
```

```python
#automatic processing

norm = em.emg_process(emg_signal=emg, order_env=3, fourier='No')
```
